## matplot example workflow

```python
# step 1: import matplot lib and get it ready for plotting in the notebook
# step 2: prepare data
# step 3: setup plot
#     fig, ax = plt.subplots(figsize=(10,10))   #width, height respectively
# step 4: ax.plot(x,y)
# step 5: ax.set(title = "Simple plot", xlabel = "X-axis", ylabel = "Yaxis")
# Step 6: save the figure in the file -> fig.savefig("image name.png")
```

```python
fig, ax = plt.subplots(figsize=(10,10))
ax.plot(x,y)
ax.set(title = "Simple plot", xlabel = "X-axis", ylabel = "Yaxis")
fig.savefig("sample-plot.png")
```

```python
# make a plot from dictionary-> bar graph
nut_butter  = {"Almond butt" : 10,"Peanut":8,"cashew butt":12}
fig,ax = plt.subplots()
ax.bar(nut_butter.keys(), nut_butter.values())
ax.set(title="kale dudh dairy", xlabel = "type", ylabel="prices")
```

```python
x= np.random.randn(1000) #hist
fig,ax = plt.subplots()
ax.hist(x);
```

# subplot can be created by 2:

```python
# opt 1:
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows = 2,ncols = 2, figsize=(10,5))
#  plot in each axes:
ax1.plot(x,x/2);
ax2.scatter(np.random.random(10),np.random.random(10));
ax3.bar(nut_butter.keys(), nut_butter.values());
ax4.hist(np.random.randn(1000));
```

```python
fig, ax = plt.subplots()
ax.scatter(x,x**2)
```

```python
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2,ncols=2,figsize=(10,5))
ax4.hist(x)
ax3.bar(x**2,x)
ax3.set(title="bar", xlabel = "x axis", ylabel = "y axis")
```

```python
# subplot plotting option 2:
fig, ax = plt.subplots(nrows = 2,ncols = 2, figsize=(10,5))
#  plot in each axes:
ax[0,0].plot(x,x/2);
ax[0,1].scatter(np.random.random(10),np.random.random(10));
ax[1,0].bar(nut_butter.keys(), nut_butter.values());
ax[1,1].hist(np.random.randn(1000));
```

```python
heart= pd.read_csv("heart-disease.csv")
heart
```

```python
heart["age"].plot.hist(bins=10);
```

```python
heart.head()
```

```python
heart.plot.hist(subplots=True, figsize=(10,30));
```

```python
over_50 = heart[heart["age"]>50]
over_50
```

```python
# pyplot method
over_50.plot(x="age", y="chol",kind="scatter", c = "target");
```

```python
# oo method
fig,ax = plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter",x="age",y="chol",c="target",ax=ax);
# ax.set_xlim([45,100])
```

```python
fig, ax = plt.subplots(figsize=(10,6))
scatter = ax.scatter(x=over_50["age"],y=over_50["chol"],c=over_50['target'])
```

```python
fig, ax = plt.subplots(figsize=(10,6))
scat = ax.scatter(x=over_50["age"], y=over_50["chol"], c=over_50["target"])
ax.set(title="Heart disease", xlabel= "Age",ylabel="Cholestrol")
ax.legend(*scatter.legend_elements(),title="Target")
ax.axhline(over_50["chol"].mean(), linestyle='--')
```

```python
over_50
```

```python
fig, ax= plt.subplots(figsize=(10,6))
scat = ax.scatter(x=over_50["thalach"],y=over_50["chol"],c=over_50["target"])
ax.legend(*scatter.legend_elements(),title="Target")
ax.axhline(over_50["chol"].mean(),linestyle='--')
```

```python
# subplot of age,chol and thalach
fig, (ax0,ax1) = plt.subplots(nrows=2,ncols=1, figsize=(10,10), sharex=True)
scatter = ax0.scatter(x=over_50["age"], y=over_50["chol"], c=over_50["target"])
ax0.set(title="Heart disease and chol levels",ylabel="Cholestrol")
ax0.legend(*scatter.legend_elements(), title="Target")
ax0.axhline(over_50["chol"].mean(), linestyle='--')

scat = ax1.scatter(x=over_50['age'],y=over_50['thalach'],c=over_50['target'])
ax1.set(title="Heart disease and Heart rate", xlabel= "Age",ylabel="Max heart rate")
ax1.legend(*scatter.legend_elements(),title="Target")
ax1.axhline(over_50["thalach"].mean(), linestyle='--')

fig.suptitle("Heart disease Analysis", fontsize=16, fontweight="bold")
```
